# Employee CRUD Operations

This notebook demonstrates Create, Read, Update, and Delete (CRUD) operations for the Employee model.

## Employee Model Fields

- **id** (Integer, Primary Key) - Auto-generated
- **fullname** (String, max 50 chars) - Required
- **job_title** (String, max 30 chars) - Optional
- **age** (Integer) - Optional, must be >= 1

```python
# Setup imports
import sys
sys.path.insert(0, '..')

from teamup.database import get_session
from teamup.repository import EmployeeRepository

print("✓ Imports ready")
```

## Create - Adding New Employees

Let's create some employees with different field combinations.

```python
# Create employee with all fields
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Full information
    emp1 = repo.create(
        fullname="Alice Johnson",
        job_title="Senior Developer",
        age=32
    )
    print(f"Created: {emp1}")
    
    # Only required fields
    emp2 = repo.create(fullname="Bob Smith")
    print(f"Created: {emp2}")
    
    # Name and age only
    emp3 = repo.create(fullname="Carol White", age=28)
    print(f"Created: {emp3}")
```

## Read - Retrieving Employees

Multiple ways to retrieve employee records.

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Get all employees
    all_employees = repo.get_all()
    print(f"\nTotal employees: {len(all_employees)}")
    for emp in all_employees[:5]:  # Show first 5
        print(f"  - {emp}")
```

```python
# Get by ID
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Get first employee's ID
    first_emp = repo.get_all()[0]
    emp_id = first_emp.id
    
    # Retrieve by ID
    retrieved = repo.get_by_id(emp_id)
    print(f"\nRetrieved by ID {emp_id}: {retrieved}")
```

## Search - Finding Employees

Use search methods to find specific employees.

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Search by name (case-insensitive, partial match)
    results = repo.find_by_name("Alice")
    print(f"\nFound {len(results)} employee(s) with 'Alice' in name:")
    for emp in results:
        print(f"  - {emp}")
```

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Search by job title
    developers = repo.find_by_job_title("Developer")
    print(f"\nFound {len(developers)} developer(s):")
    for emp in developers:
        print(f"  - {emp}")
```

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Search by age range
    young_employees = repo.find_by_age_range(min_age=25, max_age=35)
    print(f"\nEmployees aged 25-35: {len(young_employees)}")
    for emp in young_employees[:5]:  # Show first 5
        print(f"  - {emp.fullname}, age {emp.age}")
```

## Update - Modifying Employee Records

Update existing employee information.

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Find Alice
    alice = repo.find_by_name("Alice Johnson")[0]
    print(f"Before update: {alice}")
    
    # Update all fields
    updated = repo.update(
        alice.id,
        fullname="Alice Johnson-Smith",
        job_title="Principal Developer",
        age=33
    )
    print(f"After update: {updated}")
```

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Partial update - only change job title
    bob = repo.find_by_name("Bob Smith")[0]
    print(f"\nBefore: {bob}")
    
    updated = repo.update(bob.id, job_title="Junior Developer")
    print(f"After: {updated}")
```

## Delete - Removing Employees

Delete employee records from the database.

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    
    # Find Carol
    carol = repo.find_by_name("Carol White")[0]
    carol_id = carol.id
    print(f"Deleting: {carol}")
    
    # Delete the employee
    success = repo.delete(carol_id)
    print(f"Deleted: {success}")
    
    # Verify deletion
    not_found = repo.get_by_id(carol_id)
    print(f"Employee still exists: {not_found is not None}")
```

## Count - Get Total Records

Check how many employees exist.

```python
with get_session() as session:
    repo = EmployeeRepository(session)
    total = repo.count()
    print(f"Total employees in database: {total}")
```

## Using Faker for Test Data

Generate realistic test data with the Faker library.

```python
from faker import Faker

fake = Faker()

with get_session() as session:
    repo = EmployeeRepository(session)
    
    print("Creating 5 employees with Faker data:")
    for i in range(5):
        emp = repo.create(
            fullname=fake.name()[:50],  # Ensure max length
            job_title=fake.job()[:30],   # Ensure max length
            age=fake.random_int(min=22, max=65)
        )
        print(f"  {i+1}. {emp}")
```
